```python
#importamos las librerias 
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#carga de archivo 
data=pd.read_csv('Detalle precios y productos fabricados 2022_sin_nulos.csv')
```

```python
#eliminamos primera columna innecesaria  
data= data.drop(['Unnamed: 0'], axis=1)
```

```python
data.info()
```

```python
#Creo 2 dataframe para poder procesar los outliers
cuantitativas= data.iloc[ : , [4, 7, 8, 9, 10, 11, 12, 13] ]
cualitativas= data.iloc[ : , [0, 1, 2, 3, 5, 6] ]
```

```python
cuantitativas.info()
```

```python
#realizamos diagrama de caja o bigote 
fig = plt.figure(figsize= (15, 8))
cuantitativas.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show() #dibujamos el diagrama
```

```python
#Método aplicando desviación estándar. Encuentro los valores extremos
y=cuantitativas
Lim_Superior = y.mean() + 3*y.std()
Lim_Inferior = y.mean() - 3*y.std()
print("Limite superioir permitido", Lim_Superior)
print("Limite inferior permitido", Lim_Inferior)
```

```python
#Obtenemos datos y los outliers se convierten en nulos en el DataFrame
data2=cuantitativas[(y<=Lim_Superior)&(y>=Lim_Inferior)]
data2
```

```python
#corroboramos valores nulos (atipicos) del nuevo dataframe 
vna=data2.isnull().sum()
vna
```

```python
#ahora reemplazaremos valores nulos con mean (promedio) y copiamos en un nuevo dataframe 
data_clean = data2.copy()
data_clean = data_clean.fillna(round(data2.mean(), 1))
data_clean
```

```python
#corroboramos valores nulos del dataframe limpio 
vna = data_clean.isnull().sum()
vna
```

*cuantiles*

```python
#Método aplicando cuartiles. Encuentro 0.25 y 0.75
y=cuantitativas

percentiles25=y.quantile(0.25) #Q1
percentiles75=y.quantile(0.75) #Q3
iqr= percentiles75 - percentiles25

Limite_Superior_iqr= percentiles75 + 1.5*iqr
Limite_Inferior_iqr = percentiles25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#obtenemos datos limpios del Dataframe 
data3_iqr=cuantitativas[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data3_iqr
```

```python
#Corroboramos valores nulos 
valores_nulos = data3_iqr.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe 
data4_iqr = data3_iqr.copy()
data4_iqr = data4_iqr.fillna(round(data3_iqr.mean(),1))
data4_iqr
```

```python
#unimos el dataframe cuantitativo limpio con el dataframe cualitativo 
Datos_Limpios = pd.concat([cualitativas, data4_iqr], axis=1)
Datos_Limpios
```

```python
valores_nulos=Datos_Limpios.isnull().sum()
valores_nulos
```

```python
#Convertir DataFrame a csv
Datos_Limpios.to_csv("Datos_Limpios_detalles_precios.csv")
```
